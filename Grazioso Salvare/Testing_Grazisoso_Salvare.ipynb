```python
# Import the AnimalShelter class
from Grazioso_Salvare import AnimalShelter

# Instantiate the AnimalShelter class
shelter = AnimalShelter()

# Define the sample animal data dictionary
animal_data = {
    "rec_num": 1,
    "age_upon_outcome": "4 years",
    "animal_id": "C746874",
    "animal_type": "Cat",
    "breed": "Domestic Shorthair Mix",
    "color": "Black/White",
    "date_of_birth": "2014-04-10",
    "datetime": "2017-04-11 09:00:00",
    "monthyear": "2017-04-11T09:00:00",
    "name": "",
    "outcome_subtype": "SCRP",
    "outcome_type": "Transfer",
    "sex_upon_outcome": "Neutered Male",
    "location_lat": 30.5066578739455,
    "location_long": -97.3408780722188,
    "age_upon_outcome_in_weeks": 156.767857142857
}

# Test the create method
try:
    create_result = shelter.create(animal_data)
    print("Create result:", create_result)
except Exception as e:
    print(f"Error: {e}")

# Define a query to read the animal data
query = {"animal_id": "A746874"}

# Test the read method
try:
    animals = shelter.read(query)
    if animals:
        print("Animals found:")
        for animal in animals:
            print(animal)
    else:
        print("No animals found.")
except Exception as e:
    print(f"Error: {e}")
    
# Define the update data
update_query = {"animal_id": "C746874"}
new_values = {"$set": {"name": "Whiskers"}}

# Test the update method
try:
    update_result = shelter.update(update_query, new_values)
    print("Update result:", update_result)
except Exception as e:
    print(f"Error: {e}")

# Test the read method again to verify the update
try:
    updated_animals = shelter.read(update_query)
    if updated_animals:
        print("Updated animals found:")
        for animal in updated_animals:
            print(animal)
    else:
        print("No updated animals found.")
except Exception as e:
    print(f"Error: {e}")

# Define a query to delete the animal data
delete_query = {"animal_id": "C746874"}

# Test the delete method
try:
    delete_result = shelter.delete(delete_query)
    print("Delete result:", delete_result)
except Exception as e:
    print(f"Error: {e}")

# Test the read method again to verify the deletion
try:
    deleted_animals = shelter.read(delete_query)
    if deleted_animals:
        print("Animals found after deletion (should be empty):")
        for animal in deleted_animals:
            print(animal)
    else:
        print("No animals found after deletion.")
except Exception as e:
    print(f"Error: {e}")
```
